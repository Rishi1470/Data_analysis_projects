# Risk Analysis

Now, we are going to perform the Risk analysis on both the dataset. Those are daily returns, volatility, Value at Risk (VaR), and the Sharpe ratio.

```python
dataset_1['Moving Average'] = dataset_1['Close'].rolling(window=5).mean()
plt.figure(figsize=(15, 5))
plt.plot(dataset_1['Date'], dataset_1['Close'], color='blue', label='Closing Price')
plt.plot(dataset_1['Date'], dataset_1['Moving Average'], color='red', label='Moving Average')
plt.xticks(rotation=90)
plt.suptitle('Moving Average of Closing Prices in Nifty',fontsize=20)
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
```

```python
dataset_2['Moving Average'] = dataset_2['Close'].rolling(window=5).mean()
plt.figure(figsize=(15, 5))
plt.plot(dataset_2['Date'], dataset_2['Close'], color='blue', label='Closing Price')
plt.plot(dataset_2['Date'], dataset_2['Moving Average'], color='red', label='Moving Average')
plt.xticks(rotation=90)
plt.suptitle('Moving Average of Closing Prices in Sensex',fontsize=20)
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
```

```python
nifty_daily_returns = dataset_1['Open'].pct_change()
sensex_daily_returns = dataset_2['Open'].pct_change()
print("Nifty Daily Returns:", nifty_daily_returns)
print("Sensex Daily Returns:", sensex_daily_returns)
```

```python
nifty_volatility = nifty_daily_returns.std()
sensex_volatility = sensex_daily_returns.std()

print("Nifty Volatility:", nifty_volatility)
print("Sensex Volatility:", sensex_volatility)
```

```python
nifty_var = nifty_daily_returns.mean() - (nifty_volatility * 1.96)
sensex_var = sensex_daily_returns.mean() - (sensex_volatility * 1.96)

print("Nifty VaR:", nifty_var)
print("Sensex VaR:", sensex_var)
```

```python
nifty_sharpe_ratio = nifty_daily_returns.mean() / nifty_volatility
sensex_sharpe_ratio = sensex_daily_returns.mean() / sensex_volatility

print("Nifty Sharpe Ratio:", nifty_sharpe_ratio)
print("Sensex Sharpe Ratio:", sensex_sharpe_ratio)
```
